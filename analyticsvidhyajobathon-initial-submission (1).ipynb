```python
train_df['Holding_Policy_Duration'].replace({"14+": "15.0"}, inplace=True)
train_df['Holding_Policy_Duration']=train_df['Holding_Policy_Duration'].astype(float)
train_df['Holding_Policy_Duration']=train_df['Holding_Policy_Duration'].fillna(train_df['Holding_Policy_Duration'].mean())
train_df['Holding_Policy_Type']=train_df['Holding_Policy_Type'].fillna(train_df['Holding_Policy_Type'].mean())
train_df['Health Indicator']=train_df['Health Indicator'].fillna('X10')
```

```python
train_df['Accomodation_Type'] =train_df['Accomodation_Type'].astype('category').cat.codes
train_df['Reco_Insurance_Type'] =train_df['Reco_Insurance_Type'].astype('category').cat.codes
train_df['Health Indicator'] =train_df['Health Indicator'].astype('category').cat.codes
train_df['Is_Spouse'] =train_df['Is_Spouse'].astype('category').cat.codes
train_df['City_Code'] =train_df['City_Code'].astype('category').cat.codes
```

```python
y = train_df['Response']
train_df.drop(['Response'], axis =1, inplace=True)
```

```python
train_df.head()
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_train_minmax=min_max_scaler.fit_transform(train_df)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_minmax, y, test_size=0.3, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt 
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
